# Equation Based Models

Lets see a system with small number of variables. We will try to understand dynamics of love between Romeo and Juliet. 

## Model of Love

 - $<0$: Hate
 - $0$  : Neutral
 - $>0$: Love
 
### The love of Romeo and Juliet
 - $R$: The love of Romeo
 - $J$: The love of Juliet

### How their love change over time?
What are the underlying mechanism that can produce oscilation of love between the two? What causes the oscilatory dynamics of love between the two?

The following rules are our answers:

> When Juliet's love increases, Romeo's love also increase!! 
$$
\frac{dR}{dt} = +aJ
$$

> When Romeo's love increases, Juliet's love decrease!!
$$
\frac{dJ}{dt} = - bR
$$

### Their Love in the next time step

$$
R = R + \frac{dR}{dt} \Delta t
$$

$$
J = J + \frac{dJ}{dt} \Delta t
$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters are just 1
a, b = 1,1
# \delta t : time interval is 0.0001 
dt = 0.0001
# Time horizon
time = np.arange(0, 10, dt)
# List for Romeo's Love
R = [1];
# List for Juliet's love
J = [-1];
# Loop over time horizon
for t in range(len(time)-1):
    dR = a * J[-1] * dt
    dJ = -b * R[-1] * dt
    
    newR = R[-1] + dR 
    newJ = J[-1] + dJ
    
    # add to the List for Romeo's Love
    R.append(newR)
    # add to the List for Romeo's Love
    J.append(newJ)
    
print(len(R))
print(len(J))
print(len(time))
```

```python
plt.figure(figsize=(10,6))
plt.plot(R,'g-',markersize=10)
plt.plot(J,'r--',markersize=10)
plt.grid(True) 
plt.ylabel('Ask'); 
plt.xlabel('t')
plt.legend(("The Love of Romeo", "The Love of  Juliet"), shadow=True, fancybox=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(R,J,'b-')
plt.plot(R[0],J[0],'o', markersize=10)
plt.plot(R[1000],J[1000],'o', markersize=10)
plt.grid(True)
plt.xlabel('Romeo'); plt.ylabel('Juliet'); 
plt.title('Phase Space of Love')
```

# Prey-Predator Dynamics

Let x(t) denotes the population of the prey species, and y(t) denotes the population of the predator species.

$$
\frac{dx}{dt} = ax - b xy
$$

$$
\frac{dy}{dt} = - cy + d xy
$$

> Note that in the absence of the predators (when y = 0), the prey population would grow exponentially. If the preys are absence (when x = 0), the predator population would decay exponentially to zero due to starvation.

```python
import numpy as np
a, b, c, d = 1,1,1,1
x, y = [0.5], [0.1];


dt = 0.001
# Time horizon
time = np.arange(0, 30, dt)

for t in time:
    dx = (a * x[-1] - b * x[-1]* y[-1]) * dt
    dy = (-c * y[-1] + d * x[-1]* y[-1]) * dt
    
    new_x = x[-1] + dx
    new_y = y[-1] + dy
    
    x.append(new_x)
    y.append(new_y)
    
print(len(x))
print(len(y))
print(len(time))
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,'g-',markersize=10)
plt.plot(y,'r--',markersize=10)
plt.grid(True) 
plt.ylabel('Ask'); 
plt.xlabel('t')
plt.legend(("x (prey)", "y (predator)"), shadow=True, fancybox=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,'b-')
plt.plot(x[0],y[0],'o', markersize=10)
plt.plot(x[1000],y[1000],'o', markersize=10)
plt.grid(True)
plt.xlabel('x (prey)'); plt.ylabel('y (predator)'); 
plt.title('Phase Space of Prey-predator')
```
