# Simple Economic Model
 - Agents initially have some amount of money
 - In each iteration, one `giver` and one `taker` is selected
 - `giver` gives some amount of money to the `taker`
 
What will be the outcome?

```python
import numpy as np # for random 
import matplotlib.pyplot as plt # for visual effects
```

```python
class Ecoman():
    def __init__(self, ID, money = 10):
        self.ID = ID
        self.budget = money
        
    def take(self, amount):
        self.budget = self.budget + amount
        
    def give(self, amount = 1):
        """control for budget, before giving money"""
        if self.budget >= amount:
            self.budget -= amount
            return amount
        return 0
    
    def give_without_control(self, amount = 1):
        """No control for budget, in order to give the amount of money"""
        self.budget -= amount
        return amount
```

```python
class Ecomania():
    def __init__(self, N = 10, CONTROL = False):
        self.control = CONTROL
        self.N = N
        self.population = [Ecoman(i) for i in range(self.N)]
        
    def exchange(self):
        taker_index, giver_index = np.random.choice(self.N, size = 2, replace = False)
        
        if self.control:
            amount = self.population[giver_index].give()
        else:
            amount = self.population[giver_index].give_without_control()
        self.population[taker_index].take(amount)
        
    def run(self, G = 100):
        # Do 100 transactions/exchange of money
        for i in range(G):
            self.exchange()
```

# Initial Virtual World

Control is True
 - Agents control their budget before giving money to other

```python
world = Ecomania(N = 100, CONTROL = True)
```

```python
budgets = [e.budget for e in world.population]
```

```python
plt.hist(budgets)
plt.xlabel('Budget of the Agent')
plt.ylabel('Number of agents')
```

## After 1 exchange

```python
world.exchange()
```

```python
budgets = [e.budget for e in world.population]
```

```python
plt.hist(budgets)
plt.xlabel('Budget of the Agent')
plt.ylabel('Number of agents')
plt.grid()
```

# After 100 exchanges

```python
world.run(G =100)
```

```python
budgets = [e.budget for e in world.population]
plt.hist(budgets, bins = 50)
plt.xlabel('Budget of the Agent')
plt.ylabel('Number of agents')
plt.grid()
```

```python
world.run(G =1000)
budgets = [e.budget for e in world.population]
plt.hist(budgets, bins = 50)
plt.xlabel('Budget of the Agent')
plt.ylabel('Number of agents')
plt.grid()
```

# Simple Economy
 - This time, agents do not control their budget
 - They give money to other, even if they dont have money (negative budget is allowed)

```python
world = Ecomania(N = 100, CONTROL = False)
```

```python
world.run(G =1000000)
budgets = [e.budget for e in world.population]
plt.hist(budgets, bins = 50)
plt.xlabel('Budget of the Agent')
plt.ylabel('Number of agents')
plt.grid()
```

# Outcome
 - Control of budget => Power law distrbution
 - without budget control => Normal distrbution
